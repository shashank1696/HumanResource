```python
import pandas as pd
import numpy as np
url="H://python//HR_comma_sep.csv"
```

```python
data=pd.read_csv(url)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data=pd.get_dummies(data)
```

```python
data.head()
```

```python
data.columns
```

```python
data=data.drop(['sales_IT','salary_low'],axis=1)
```

```python
y=data['left'] ## target
X=data.drop('left',axis=1) ## predictors
```

```python
import sklearn.model_selection as model_selection
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.20,
                                  random_state=200)
```

```python
import sklearn.tree as tree
```

```python
clf=tree.DecisionTreeClassifier(max_depth=3)
```

```python
clf=clf.fit(X_train,y_train)
```

```python
## plot the model
import pydotplus
#import os
#os.environ['PATH']+=os.pathsep+"C:/Program Files (x86)/Graphviz 2.38/bin/"
```

```python
graph=tree.export_graphviz(clf,feature_names=X_train.columns,
                          class_names=["0","1"],proportion=True,filled=True,
                          out_file=None)
```

```python
graph
```

```python
graph=pydotplus.graph_from_dot_data(graph)
```

```python
X_test.iloc[0]
```

```python
y_test_prediction=clf.predict(X_test)
```

```python
results=pd.DataFrame({'actual':y_test,'predicted':y_test_prediction})
```

```python
clf.score(X_test,y_test)
```

```python
### Create another model with max_depth 8 and check the accuracy
clf1=tree.DecisionTreeClassifier(max_depth=8)
clf1=clf1.fit(X_train,y_train)
```

```python
clf1.score(X_test,y_test)
```

```python
### Confusion Matrix
import sklearn.metrics as metrics
```

```python
y_test_prediction # model with depth of 3
y_test_prediction2=clf1.predict(X_test)
```

```python
## Confusion Matrix for model with depth of 3
metrics.confusion_matrix(y_test,y_test_prediction)
```

```python
## Model with depth 8
metrics.confusion_matrix(y_test,y_test_prediction2)
```
